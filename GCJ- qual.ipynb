```
# vestigium problem
#matrix
case = 0
T = int(input())
for i in range(T):
  matrix = list()
  N = int(input())
  case += 1
  matrix = [[int(x) for x in input().split(" ")] for y in range(N)]

  # trace- diagonal sum  
  sum = 0
  R_cnt = 0
  C_cnt = 0
  for i in range(N):
    sum += matrix[i][i]
    # repeated elements in rows
    if len(set(matrix[i])) != N: 
      R_cnt += 1
    # repeated elements in columns
    #res = list(zip(list(*matrix)[i]))
    res = [sub[i] for sub in matrix]
    if len(set(res)) != N:
      C_cnt += 1




  print(("Case #"+str(case)+": ")+ str(sum) + " " + str(R_cnt) + " " + str(C_cnt))
```

```
#problem3
class activity:
    def __init__(self, num, start, end):
        self.num = num
        self.start = start
        self.end = end
        self.who=""


def addActiviy( aListe, anActivity, who):
    ret = True
    if(len(aListe)==0):
        anActivity.who = who
        aListe.append(anActivity)
        return True
   
    lastActivity = aListe[ -1]
    if lastActivity.end > anActivity.start :
        return False
    anActivity.who = who
    aListe.append(anActivity)
   
    return True


def organize(aDay, Jamie, Cameron ):
   
    if(len(aDay)==0):
        return True
   
    a = aDay.pop(0)
   
    ret = addActiviy( Cameron, a, "C")
    if ret :
        r =organize(aDay, Jamie, Cameron )
        return r

    ret = addActiviy( Jamie, a, "J")
    if ret :
        r = organize(aDay, Jamie, Cameron )
        return r
   
    return False
   

def displayActivity( l, testCaseNum):
    s=""
    for x in l:
        s+=x.who
    print( "Case #"+str(testCaseNum)+": "+ s)

def organizeDay( aDay, testCaseNum):
    aDay.sort(key=lambda x: x.start)
   
    Jamie=[]
    Cameron=[]
    isPossible = organize(aDay, Jamie, Cameron )
   
    if(isPossible):
        l = Jamie + Cameron
        l.sort(key=lambda x: x.num)
        displayActivity( l, testCaseNum)
    else:
        print( "Case #"+str(testCaseNum)+": IMPOSSIBLE")


def main():
  A = int(input())
  testCaseNum = 0
  for i in range (A):
    n = int(input())
    day = []
    testCaseNum+=1
    for i in range (n):
      b = input().split(" ")
      a =  activity( i, int(b[0]),int(b[1]))
      day.append(a)  
    organizeDay( day, testCaseNum)

if __name__ = "__main__":
  main()

```
